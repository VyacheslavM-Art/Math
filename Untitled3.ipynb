```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
# Построение окружности радиусом 2
x=np.linspace(-2,2,200)
y=(4-x**2)**(1/2)
y1=-(4-x**2)**(1/2)
plt.plot(x,y)
plt.plot(x,y1)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```

```
# Построение элипса при b=1,a=2
x=np.linspace(-2,2,200)
y=(1-((x**2)/4))**(1/2)
y1=-(1-((x**2)/4))**(1/2)
plt.plot(x,y)
plt.plot(x,y1)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.show()
```

```
# гипербола
import decimal

xmin = -20
xmax = 20
dx = 0.1

xlist = np.around(np.arange(xmin, xmax, dx), decimals=4)

ylist = 1 / xlist

plt.plot(xlist, ylist)
plt.show()
```

```
#паралельные плоскости
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 20)
X, Y = np.meshgrid(X, Y)
Z = X+Y
Z2 = X + Y+5
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```
#плоскости второго порядка
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 2)
Y = np.arange(-50, 50, 10)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2
Z2 = 5*X**2 + 3*Y**2+7
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```
